```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from catboost import CatBoostRegressor
from sklearn.impute import KNNImputer
```

```python
train_df = pd.read_csv("/kaggle/input/neoai-2025-tricy-table-data/train_tables.csv")
test_df = pd.read_csv("/kaggle/input/neoai-2025-tricy-table-data/test_tables.csv")
submission_df = pd.read_csv("/kaggle/input/neoai-2025-tricy-table-data/sample_submission.csv")
```

```python
scaler = MinMaxScaler()
scaler1 = StandardScaler()

X_train = train_df.copy()
y_train = train_df.target
X_train = X_train.drop('target', axis=1)
X_train = pd.DataFrame(scaler1.fit_transform(X_train), columns=X_train.columns)

X_test = test_df.copy().drop('id', axis=1)
X_test = pd.DataFrame(scaler1.transform(X_test), columns=X_test.columns)


X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
train_df
```

```python
imp = KNNImputer(n_neighbors=10)
X_train = pd.DataFrame(imp.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(imp.transform(X_test), columns=X_test.columns)
```

```python
X_train['hour_sin'] = np.sin(X_train['hour'] * np.pi * 2 / 24)
X_train['hour_cos'] = np.cos(X_train['hour'] * np.pi * 2 / 24)
X_train['min_sin'] = np.sin(X_train['minute'] * np.pi * 2 / 60)
X_train['min_cos'] = np.cos(X_train['minute'] * np.pi * 2 / 60)

X_train.drop(['hour', 'minute'], axis=1, inplace=True, errors='ignore')

X_test['hour_sin'] = np.sin(X_test['hour'] * np.pi * 2 / 24)
X_test['hour_cos'] = np.cos(X_test['hour'] * np.pi * 2 / 24)
X_test['min_sin'] = np.sin(X_test['minute'] * np.pi * 2 / 60)
X_test['min_cos'] = np.cos(X_test['minute'] * np.pi * 2 / 60)

X_test.drop(['hour', 'minute'], axis=1, inplace=True, errors='ignore')
```

```python
model = CatBoostRegressor()

model.fit(X_train, y_train)
```

```python

preds = model.predict(X_test)
submission_df['target'] = preds
submission_df.to_csv('submission.csv', index=False)
```
