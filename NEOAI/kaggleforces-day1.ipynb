# Description

The year is 2026. Kaggle has fully transformed into a digital arena where the best machine learning specialists compete for medals, prize pools, and leaderboard immortality.

After decades of competitions, humanity has accumulated massive archives of Kaggle history: public and private ranks, performance trends, submission behavior, reward structures, and participant statistics across countless competitions.

Yet one question still remains unanswered:

> Can we predict who will become one of the strongest competitors in the next Featured competition?

To answer this question, you must build a Kaggle success prediction system.

For every participant in the test competitions, you need to predict a continuous score representing the probability of finishing in the top 3% of the original leaderboard.

Top-3% is computed relative to the full size of the original competition:

```text
PrivateLeaderboardRank / TotalTeams ≤ 0.03
```

Important: the test set contains only filtered users with sufficient competition history, but the target itself is always computed relative to the full original Kaggle leaderboard.

````python
import pandas as pd
from catboost import CatBoostClassifier

train = pd.read_parquet("/kaggle/input/competitions/neoai-2026-day-1-kaggleforces/train.parquet")
test = pd.read_csv("/kaggle/input/competitions/neoai-2026-day-1-kaggleforces/test.csv")

for c in ["RewardQuantity", "final_public_score", "final_private_score"]:
    train[c] = pd.to_numeric(train[c], errors="coerce")

train["target"] = (
    train["PrivateLeaderboardRank"] / train["TotalTeams"] <= 0.03
).astype(int)

train["rank_pct"] = train["PrivateLeaderboardRank"] / train["TotalTeams"]

train["gold"] = (train["Medal"] == 1).astype(int)
train["silver"] = (train["Medal"] == 2).astype(int)
train["bronze"] = (train["Medal"] == 3).astype(int)

user_features = train.groupby("UserId").agg(
    n_comp=("CompetitionId", "count"),
    top3_rate=("target", "mean"),
    top3_count=("target", "sum"),
    avg_rank_pct=("rank_pct", "mean"),
    best_rank_pct=("rank_pct", "min"),
    worst_rank_pct=("rank_pct", "max"),
    avg_public_score=("final_public_score", "mean"),
    avg_private_score=("final_private_score", "mean"),
    avg_submissions=("n_submissions", "mean"),
    total_submissions=("n_submissions", "sum"),
    golds=("gold", "sum"),
    silvers=("silver", "sum"),
    bronzes=("bronze", "sum"),
    avg_reward=("RewardQuantity", "mean"),
    max_reward=("RewardQuantity", "max"),
    avg_totalteams=("TotalTeams", "mean"),
).reset_index()

train = train.merge(user_features, on="UserId", how="left")
test = test.merge(user_features, on="UserId", how="left")

features = [
    "n_comp",
    "top3_rate",
    "top3_count",
    "avg_rank_pct",
    "best_rank_pct",
    "worst_rank_pct",
    "avg_public_score",
    "avg_private_score",
    "avg_submissions",
    "total_submissions",
    "golds",
    "silvers",
    "bronzes",
    "avg_reward",
    "max_reward",
    "avg_totalteams",
]

X_train = train[features].fillna(-1)
y_train = train["target"]

X_test = test[features].fillna(-1)

model = CatBoostClassifier(
    iterations=500,
    depth=6,
    learning_rate=0.05,
    loss_function="Logloss",
    eval_metric="AUC",
    random_seed=42,
    verbose=100,
)

model.fit(X_train, y_train)

pred = model.predict_proba(X_test)[:, 1]

submission = pd.DataFrame({
    "Id": test["CompetitionId"].astype(str) + "_" + test["UserId"].astype(str),
    "pred_score": pred,
})

submission.to_csv("submission.csv", index=False)
````
