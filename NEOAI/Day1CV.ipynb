## Overview
In the previous phase, humanity successfully recovered the lost knowledge of pet breeds. Now, we face a new challenge: **Image Segmentation**. 

You will need to train a machine learning model capable of precisely separating our furry friends from their terminal backgrounds. Return clean, segmented avatars to humanity using only the power of AI!

## Description
The year is 2056. 
Thanks to your previous efforts, humanity finally remembers which text file contains a pug and which contains a Maine Coon. However, as we dug deeper into the archives, we realized that our terminal pets are trapped. They are surrounded by legacy console clutter, ASCII-art furniture, and digital background noise. We can't use these as our console tamagotchis!

To make matters more complex, we discovered a cache of advanced "ANSI TrueColor" files. During the 'Great Console Transition,' the images weren't just converted to text - they were rendered as colored characters. 

The converter used a strict palette to convey lighting, shadows, and contours:
*   **Fill characters:** `[" ", ".", ";", "c", "o", "P", "O", "?", "@", "■"]`
*   **Edge characters:** `["|", "-", "\", "/"]`

These characters were ultimately rendered into colored image files, with each character occupying a precise 10x10 pixel block. 

Your mission is to perform **pixel-perfect (or rather, character-perfect)** segmentation to cut the pets out of their terminal backgrounds. 

## Data
Unlike the previous text-based challenge, this time you will be working with rendered image files. 

*   **train.csv**: A tabular file containing the training metadata with the following columns:
    *   `id`: The unique identifier for the image.
    *   `img`: The file path to the rendered, colored ASCII image.
    *   `label`: The file path to the ground-truth segmentation mask.
*   **test.csv**: A tabular file containing the test metadata, providing the `id` and `img` for each image you need to segment.
*   **Images**: Rendered colored ASCII art saved as standard image files (.png). Each image consists of 128x128 characters, each 10x10 pixels. The total size of each image is 1280x1280 pixels. 
*   **Masks**: Binary images corresponding to the training images, where pixel values are strictly `0` (background) and `255` (the pet).
*   **test**: A folder containing the test images for which you need to predict the binary masks.
*   **font.ttf**: The exact font file used during the original rendering process. 

## Evaluation
Participants' predictions will be evaluated using the **Dice Coefficient** metric.

````python
from torchvision import transforms
class CutiesDataset128(Dataset):
    def __init__(self, base_dir, df, subset="train"):
        self.base_dir = base_dir
        self.df = df
        self.subset = subset
        if self.subset == "train":
            self.transform = v2.Compose([
                v2.RandomHorizontalFlip(p=0.1),
                v2.ToTensor()
            ])

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = row['img']
        
        img = Image.open(self.base_dir / img_path)
        img_tensor = transforms.ToTensor()(img).float()
        
        # 2. Add a fake batch dimension: Shape [1, 3, 1280, 1280]
        img_tensor = img_tensor.unsqueeze(0)
        
        # 3. Apply a 10x10 max pool to shrink it down to 128x128
        # We use a kernel size of 10 and stride of 10 to downsample perfectly by block boundaries
        img_tensor = F.max_pool2d(img_tensor, kernel_size=10, stride=10)
        
        # 4. Remove the fake batch dimension: Shape [3, 128, 128]
        img_tensor = img_tensor.squeeze(0)
        
        # 5. Convert back to a PIL image to feed into your augmentation pipeline
        img_resized = transforms.functional.to_pil_image(img_tensor)
        
        if self.subset == "train":
            mask_path = row['label']
            mask = Image.open(self.base_dir / mask_path)
            mask_resized = mask.resize((128, 128), Image.Resampling.BOX)
            mask_resized = (np.array(mask_resized) != 0).astype(int)
            
            mask_tensor = torch.tensor(mask_resized, dtype=torch.float32).unsqueeze(0)
            mask_tv = tv_tensors.Mask(mask_tensor)
            
            img_aug, mask_aug = self.transform(img_resized, mask_tv)
            
            img_tensor = img_aug.as_subclass(torch.Tensor).float()
            mask_tensor = mask_aug.as_subclass(torch.Tensor).float()
            
            return img_tensor, mask_tensor
        elif self.subset == "val":
            img_tensor = v2.ToTensor()(img_resized)
            mask = v2.ToTensor()(Image.open(self.base_dir / row["label"]))
            return img_tensor.float(), mask.float()
        elif self.subset == "test":
            img_tensor = v2.ToTensor()(img_resized)
            return img_tensor.float()
````

````python
import torch.nn.functional as F
class BCEDiceLoss(nn.Module):
    def __init__(self, bce_weight=1.0, dice_weight=1.0, smooth=1e-5):
        super(BCEDiceLoss, self).__init__()
        self.bce_weight = bce_weight
        self.dice_weight = dice_weight
        self.smooth = smooth

    def forward(self, inputs, targets):
        targets = targets.float()
        
        bce_loss = F.binary_cross_entropy_with_logits(inputs, targets)
        
        inputs_prob = torch.sigmoid(inputs)
        
        inputs_flat = inputs_prob.view(-1)
        targets_flat = targets.view(-1)
        
        intersection = (inputs_flat * targets_flat).sum()
        union = inputs_flat.sum() + targets_flat.sum()
        
        dice_score = (2. * intersection + self.smooth) / (union + self.smooth)
        dice_loss = 1.0 - dice_score
        
        total_loss = (self.bce_weight * bce_loss) + (self.dice_weight * dice_loss)
        
        return total_loss
````

````python
class DoubleConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.net(x)


class CatboostSegmentator(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super().__init__()

        self.e1 = DoubleConv(in_channels, 32)
        self.e2 = DoubleConv(32, 64)
        self.e3 = DoubleConv(64, 128)

        self.pool = nn.MaxPool2d(2)

        self.d2 = DoubleConv(128 + 64, 64)
        self.d1 = DoubleConv(64 + 32, 32)

        self.out_conv = nn.Conv2d(32, out_channels, 1)

    def forward(self, x):
        x1 = self.e1(x)                 # 128x128
        x2 = self.e2(self.pool(x1))     # 64x64
        x3 = self.e3(self.pool(x2))     # 32x32

        y = F.interpolate(x3, size=x2.shape[2:], mode="bilinear", align_corners=False)
        y = torch.cat([y, x2], dim=1)
        y = self.d2(y)

        y = F.interpolate(y, size=x1.shape[2:], mode="bilinear", align_corners=False)
        y = torch.cat([y, x1], dim=1)
        y = self.d1(y)

        return self.out_conv(y)
````

````python
def ink_mask_from_image_fast_better(img_path, block=10, eps_strong=10, eps_weak=3):
    img = np.array(Image.open(img_path).convert("RGB"), dtype=np.int16)
    H, W = img.shape[:2]

    h, w = H // block, W // block
    cells = img.reshape(h, block, w, block, 3).transpose(0, 2, 1, 3, 4)

    bg = np.median(cells, axis=(2, 3), keepdims=True)
    diff = np.abs(cells - bg).sum(axis=4)

    strong = diff > eps_strong
    weak = diff > eps_weak

    pad = np.pad(strong, ((0, 0), (0, 0), (1, 1), (1, 1)), constant_values=False)

    near_strong = (
        pad[:, :, :-2, :-2] | pad[:, :, :-2, 1:-1] | pad[:, :, :-2, 2:] |
        pad[:, :, 1:-1, :-2] | pad[:, :, 1:-1, 1:-1] | pad[:, :, 1:-1, 2:] |
        pad[:, :, 2:, :-2] | pad[:, :, 2:, 1:-1] | pad[:, :, 2:, 2:]
    )

    ink_cells = weak & near_strong
    return ink_cells.transpose(0, 2, 1, 3).reshape(H, W)
````

````python
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=4)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = CatboostSegmentator().to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = BCEDiceLoss()

epochs = 10
for epoch in range(epochs):
    model.train()
    train_loss = 0
    for imgs, masks in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} Train"):
        imgs, masks = imgs.to(device), masks.to(device)
        
        optimizer.zero_grad()
        
        preds = model(imgs)
        loss = criterion(preds, masks)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        
    print(f"Epoch {epoch+1} Loss: {train_loss/len(train_loader):.4f}")
    
    model.eval()
    
    val_dice = 0.0

    batch_start = 0

    with torch.no_grad():
        for imgs, masks in tqdm(val_loader, desc="Validation"):
            imgs = imgs.to(device)
            preds = model(imgs)
            preds_prob = torch.sigmoid(preds).cpu().numpy()
            
            for i in range(preds_prob.shape[0]):
                row = val.iloc[batch_start + i]
                img_path = base_dir / row["img"]
    
                pred_128 = preds_prob[i, 0]
                pred_full = cv2.resize(pred_128, (1280, 1280), interpolation=cv2.INTER_NEAREST)
    
                ink = ink_mask_from_image_fast_better(img_path, eps_strong=10, eps_weak=3)
    
                pred_binary = ((pred_full > 0.5) & ink).astype(np.float32)
                orig_mask = (masks[i].cpu().numpy() > 0).astype(np.float32)
    
                intersection = (pred_binary * orig_mask).sum()
                area_sum = pred_binary.sum() + orig_mask.sum()
                dice = (2 * intersection) / area_sum if area_sum > 0 else 1.0
    
                val_dice += dice
    
            batch_start += preds_prob.shape[0]

    avg_dice = val_dice / len(val_dataset)
    
    plt.figure(figsize=(15, 5))
    
    last_img = np.array(imgs[-1].permute((1, 2, 0)).cpu())
        
    plt.subplot(1, 2, 1)
    plt.imshow(last_img)
    plt.title("Image")
    
    plt.subplot(1, 2, 2)
    plt.imshow((pred_full > 0.5).astype(int))
    plt.title("Pred")
    plt.plot()
    
    print(f"--- Epoch {epoch+1} Results ---")
    print(f"Standard | Val Dice: {avg_dice:.4f}")
    print("-" * 30)
````

````python
def rle_encode(mask):
    """
    Encodes a binary mask (numpy array) to Kaggle-style Run-Length Encoding.
    Mask should be 0 for background and >0 for foreground.
    Uses Fortran order (column-major) as is standard for Kaggle.
    """
    pixels = (mask.flatten(order='F') > 0).astype(np.uint8)
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return ' '.join(str(x) for x in runs)

def rle_decode(mask_rle, shape):
    """
    Decodes Kaggle-style RLE string to a binary mask (numpy array).
    Returns mask with shape `shape` and values 0 and 255.
    """
    if not isinstance(mask_rle, str) or mask_rle == '':
        return np.zeros(shape, dtype=np.uint8)
        
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    
    img = np.zeros(shape[0] * shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 255
        
    return img.reshape(shape, order='F')
````

````python
test_df = pd.read_csv(base_dir / "test.csv")
````

````python
test_dataset = CutiesDataset128(base_dir, test_df, subset="test")
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False, num_workers=4)
````

````python
model.eval()

rles = []
ptr = 0

with torch.no_grad():
    for imgs in tqdm(test_loader, desc="Test inference"):
        imgs = imgs.to(device)
        preds = model(imgs)
        preds_prob = torch.sigmoid(preds).cpu().numpy()

        for i in range(preds_prob.shape[0]):
            row = test_df.iloc[ptr]
            img_path = base_dir / row["img"]

            pred_128 = preds_prob[i, 0]
            pred_full = cv2.resize(pred_128, (1280, 1280), interpolation=cv2.INTER_NEAREST)

            pred_binary = pred_full > 0.5

            ink = ink_mask_from_image_fast_better(img_path, eps_strong=10, eps_weak=3)
            pred_binary = (pred_binary & ink).astype(np.uint8)

            rles.append(rle_encode(pred_binary))
            ptr += 1
````

````python
submission = pd.DataFrame({
    "id": test_df["id"],
    "mask_rle": rles
})

submission.to_csv("submission.csv.gz", index=None, compression='gzip')
````
