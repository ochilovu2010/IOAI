```python
#https://judge.nitro-ai.org/competitions/cpfed/ai-olymp-2026-second-round/1/view
import argparse
import os
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score
```

```python
train = pd.read_csv("/content/train_data.csv")
test = pd.read_csv("/content/test_data.csv")
sample = pd.read_csv("/content/sample_output.csv")
```

```python
scaler = StandardScaler()
tresholds = [0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4]
scores = []
for x in tresholds:

  model = AgglomerativeClustering(
      n_clusters = None,
      distance_threshold =x,
      metric = 'cosine',
      linkage = 'average'
  )
  X = scaler.fit_transform(train.drop(columns=['submission_id', 'author_id']))

  labels = model.fit_predict(X)
  score = adjusted_rand_score(labels, train['author_id'])
  scores.append(score)
```

```python
print(np.argmax(np.array(scores)))
```
