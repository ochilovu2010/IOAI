```python
import random
import nltk
from nltk.probability import ConditionalFreqDist
from nltk.tokenize import word_tokenize
from nltk import ngrams
```

```python
corpus = "The quick brown fox jumps over the lazy dog. The lazy dog barks loudly at the fox."
tokenized = word_tokenize(corpus.lower())
tokenized
```

```python
bigrams = list(ngrams(tokenized, 2))
print(bigrams)
```

```python
markov_chain = ConditionalFreqDist(bigrams)
```

```python
current = 'the'
outputs = [current]


for _ in range(10):
  choices = list(markov_chain[current].keys())
  if not choices:
    break
  weights = [markov_chain[current][word] for word in choices]
  randomed = random.choices(choices, weights = weights)[0]
  outputs.append(randomed)
  current = randomed
```

```python
print("".join(word + " " for word in outputs))
```
