```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
batch_size = 1
seq_length = 10
d_model = 512
i = torch.arange(0, 512, 2)
```

```python
index = 2*i/d_model
print(index.shape)
```

```python
input = torch.randn(batch_size, seq_length, d_model)
```

```python
input.shape
```

```python
denominator = torch.pow(10000, index)
denominator.shape
```

```python
positional_encoding = input
pos = torch.arange(0, seq_length)
encodings = (pos.unsqueeze(1)/denominator.unsqueeze(0))
positional_encoding = torch.cat((encodings.sin(), encodings.cos()), dim =-1).reshape(input.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(positional_encoding[0], aspect='auto')
plt.show()
```

```python
plt.plot(positional_encoding[0][9])
plt.show()
```

```python
output = positional_encoding + input
```
