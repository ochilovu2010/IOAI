```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
batch_size = 1
seq_length = 4
embedding_dim = 512
```

```python
input = torch.randn(batch_size, seq_length, embedding_dim)
print(input.shape)
```

```python
qkv_attention = torch.randn(batch_size, seq_length, embedding_dim *3)
```

```python
num_heads = 8
qkv_heads = qkv_attention.reshape(batch_size, seq_length, num_heads, embedding_dim * 3 // num_heads)
print(qkv_heads.shape)
```

```python
qkv_heads = qkv_heads.permute(0, 2, 1, 3)
print(qkv_heads.shape)
```

```python
query, key, value = qkv_heads.chunk(3, dim = -1)
```

```python
query.shape, key.shape, value.shape
```

```python
attention = query @ key.transpose(2, 3)
```

```python
attention.shape
```

```python
mask = torch.full(attention.size(), float('-inf'))
mask = torch.triu(mask, diagonal = 1)
attention = attention + mask
```

```python
attention = attention.softmax(dim = -1) # should be row wise
```

```python
values = attention@value
values.shape
```

```python
values =values.reshape(batch_size, seq_length, embedding_dim)
```

```python
values.shape
```

```python
print((input+values).shape)
print(input+values)
```
